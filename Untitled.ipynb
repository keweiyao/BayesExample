```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assume a nucleus has a three-dimensional Gaussian-like density functino
# rho(x, y, z) = (2*pi*R^2)^{-3/2} * exp(-.5*r^2/R^2)
# Thickness, TA(x, y) = 1/(2*pi*R^2) * exp(-.5*(x^2+y^2)/R^2)
A = 208
R = 1.12 * A**(1./3.) #(fm=10^{-15} m)

def TA(x, y):
    return A*np.exp(-.5*(x**2+y**2)/R**2) / (2*np.pi*R**2)

x = np.linspace(-2*R,2*R, 100)
X, Y = np.meshgrid(x,x)
plt.imshow(TA(X,Y), extent=[-2*R,2*R,-2*R, 2*R])
```

```python
# overlap function
# TAA(b) = integrate dx dy TA(x-b/2,y) * TA(x+b/2,y)
def TAA(b):
    return A**2/(4*np.pi*R**2)*np.exp(-b**2/(4*R**2))

b = np.linspace(0, 20*R, 100)
plt.plot(b, TAA(b))
plt.ylabel("Overlap [1/fm^2]")
```

```python
# The probablity for two nucleus to collide
sigmaNN = 7.0 #fm^2
Prob_miss = np.exp(-sigmaNN*TAA(b))
Prob_interact = 1. - Prob_miss
plt.plot(b, Prob_miss, label="prob to miss")
plt.plot(b, Prob_interact, label="prob to interact")
```

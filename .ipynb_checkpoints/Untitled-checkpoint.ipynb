```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Assume a nucleus has a three-dimensional Gaussian-like density functino
# rho(x, y, z) = (2*pi*R^2)^{-3/2} * exp(-.5*r^2/R^2)
# Thickness, TA(x, y) = 1/(2*pi*R^2) * exp(-.5*(x^2+y^2)/R^2)
A = 208
R = 1.12 * A**(1./3.) #(fm=10^{-15} m)

def TA(x, y):
    return A*np.exp(-.5*(x**2+y**2)/R**2) / (2*np.pi*R**2)

x = np.linspace(-2*R,2*R, 100)
X, Y = np.meshgrid(x,x)
plt.imshow(TA(X,Y), extent=[-2*R,2*R,-2*R, 2*R])
```

```python
# overlap function
# TAA(b) = integrate dx dy TA(x-b/2,y) * TA(x+b/2,y)
def TAA(b):
    return A**2/(4*np.pi*R**2)*np.exp(-b**2/(4*R**2))

b = np.linspace(0, 20*R, 100)
plt.plot(b, TAA(b))
plt.ylabel("Overlap [1/fm^2]")
```

```python
# The probablity for two nucleus to collide
sigmaNN = 7.0 #fm^2
Prob_miss = np.exp(-sigmaNN*TAA(b))
Prob_interact = 1. - Prob_miss
plt.plot(b, Prob_miss, label="prob to miss")
plt.plot(b, Prob_interact, label="prob to interact")
```

```python
# Gamma v.s. Beta distribution
from scipy.special import gamma, beta
@np.vectorize
def old(x, xavg, a):
    if x<0:
        return 0.
    return a**a/gamma(a) * (x/xavg)**(a-1) * np.exp(-a*x/xavg) / xavg

@np.vectorize
def new(x, xavg, a):
    if (x<0) or (x>1):
        return 0.
    A = np.copy(a)
    B = A * (1-xavg)/xavg
    return x**(A-1) *(1.-x)**(B-1) / beta(A, B)
```

```python
plt.figure(figsize=(6,5))
x = np.linspace(.01,2,100)
xavg = 0.15
for a, lw in zip([0.1,1, 10], [1,2,3]):
    plt.plot(x, old(x, xavg, a), 'k--', label='$\Gamma$-fluct, $a={}$'.format(a), lw=lw)
for a, lw in zip([0.1, 1, 10], [1,2,3]):
    plt.plot(x, new(x, xavg, a), 'b-', label='$B$-fluct, $a={}$'.format(a), lw=lw)
plt.semilogy()
plt.legend(ncol=2, fontsize=12)
plt.xlabel(r"$x=\Delta E/E_0$", fontsize=15)
plt.title(r"$\langle x \rangle = 0.15$", fontsize=15)
plt.ylabel(r"$P(x)$", fontsize=15)
plt.ylim(1e-5,1e3)
plt.savefig("NewParametrization.png",dpi=300
           )
```

## A simple model to understand the typical features of jet energy loss

### First, model the partonic content of the jet in the medium
Assume the partonic fragmentation function of a jet in the vacuum is $D(x) = x\frac{dN}{dx} \approx \frac{1}{x^{\alpha}}$. 
In the medium, we first consider induced gluon radiaiton, whose x-spectra is softer than that in the vacuum, so that
$$ D_M(x) \propto \frac{1}{x^{\alpha}} \left(1 + \frac{\alpha_s n}{\max\{1, [x p_T^{\rm jet}/\omega_{\rm BH}]^\beta \}} \right) \\
\int_{x_{\rm min}}^1 D_M(x) dx = 1$$,
where $n$ is an effective opacity, $\omega_{\rm BH}\sim 3T$ is the Bethe-Heitler energy for incoherent radiaiton. For $x p_T^{\rm jet} > \omega_{\rm BH}$ the induced radiation spectrum is controled by an extra power $1/x^\beta$ relative to vacuum radiation.

### Apply elastic energy loss (they cause most energy loss to large angle) to all the partons in jet
Elastic energy loss of parton is almost independent of parton energy. I assume it to be $\Delta E_p = n \alpha_s^2 \omega_{\rm BH}$, 
then the in-medium energy loss that is transfered to very large angle is then
$$\Delta E_{\rm jet} = \int_{x_{\rm min}}^{1} D_M(x) \max\left\{1, \frac{\Delta E_p}{x p_T^{\rm jet}}  \right\} dx$$

```python
from scipy.integrate import quad

@np.vectorize
def alpha_s(Q2):
    return 4*np.pi/9/np.log(Q2/0.04)

@np.vectorize
def Simple_Eloss(pTjet, alpha, n, beta, omegaBH, elastic=False):
    def dE0dx(lnx):
        x = np.exp(lnx)
        factor = 1 if elastic else (1+alpha_s(omegaBH**2)*n/(np.max([1,x*pTjet/omegaBH]))**beta) 
        return x/x**alpha*factor
    def ddEdx(lnx):
        x = np.exp(lnx)
        return dE0dx(lnx) * np.min([1, .5*alpha_s(omegaBH**2)**2*n*omegaBH/(x*pTjet)])
    A = quad(ddEdx, np.log(.3/pTjet), 1)[0]
    B = quad(dE0dx, np.log(.3/pTjet), 1)[0]
    return A/B
```

### Form of energy loss
Depending on the power exponent $\alpha$ for vacuum radiation, and $\alpha+\beta$ for medium-induced radiations (LPM region), the shape of the $\Delta E/E$ v.s. $E$ varies like a power low, as shown below.  Extra dependneces on $p_T$ can be captured by a $\ln^d p_T$ term.

This motivates the following paramterization $\Delta E = b E^c \ln^d E$, when $\ln E \gg 1$

```python
pTjet = np.exp(np.linspace(np.log(10), np.log(1000), 15))
fig, axes = plt.subplots(1,3,figsize=(15,5), sharex=True, sharey=True)
omegaBH = 1
from scipy.optimize import curve_fit
def template(lnx, a, b, c):
    x = np.exp(lnx)
    y = a * x**b * lnx**c
    return np.log(y)
for alpha, ax in zip([.7,1,1.3], axes):
    for beta, lw in zip([0,1],['-','--']):
        for n, c in zip([.5,1,1.5], ['r','b']):
            X, Y = pTjet, Simple_Eloss(pTjet, alpha, n, beta, omegaBH)
            ax.plot(X, Y, 'o', color=c)
            p,_ = curve_fit(template, np.log(X), np.log(Y))
            ax.plot(X, np.exp(template(np.log(X), *p)),lw, color=c)
for ax in axes:
    ax.semilogx()
    ax.set_xlabel(r"$p_T$", fontsize=15)
    ax.set_ylabel(r"$\langle \delta p_T\rangle/p_T$", fontsize=15)
    ax.semilogy()
```

# The resulting $R_{AA}$ of this simple model 

```python
pTjet = np.exp(np.linspace(np.log(20), np.log(1000), 15))
fig, axes = plt.subplots(1,3,figsize=(15,5), sharex=True, sharey=True)
omegaBH = 1
from scipy.optimize import curve_fit
def template(lnx, a, b, c):
    x = np.exp(lnx)
    y = a * x**b * lnx**c
    return np.log(y)
for alpha, ax in zip([.7,1,1.3], axes):
    
    for n, c in zip([.5,1,1.5], ['r','g','b']):
        
        X, Y1 = pTjet, Simple_Eloss(pTjet, alpha, n, 1, omegaBH, elastic=True)
        X, Y2 = pTjet, Simple_Eloss(pTjet, alpha, n, 1, omegaBH, elastic=False)
        ax.fill_between(X, 1-6*Y1, 1-6*Y2, color=c, label=r'$n={}$, vary $0<\beta<1$'.format(n), alpha=.5)
    ax.set_title(r"$\alpha={}$".format(alpha), fontsize=15)
    
    ax.semilogx()
    ax.set_ylim(0,1)
    ax.set_ylabel(r"$R_{AA}$", fontsize=15)
    ax.set_xlabel(r"$p_T$", fontsize=15)
```
